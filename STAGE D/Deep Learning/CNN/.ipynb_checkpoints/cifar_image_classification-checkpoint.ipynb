```python
import tensorflow as tf 
from tensorflow import keras 
from keras import datasets
from keras.layers import Dense, Flatten
from keras.models import Sequential
from keras.utils import to_categorical

import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
(train_imgs, train_labels), (test_imgs, test_labels) = datasets.cifar10.load_data()
```

```python
train_imgs.shape
```

```python
# plot the images 
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.imshow(train_imgs[88])
plt.title(classes[train_labels.reshape(-1,)[88]]);
```

```python
# make an image plotting function 
def plot_img(index):
    plt.imshow(train_imgs[index])
    plt.title(classes[train_labels.reshape(-1,)[index]]);
```

```python
plot_img(99)
```

```python
# normalising the data
X_train = train_imgs.astype(float) / 255
X_test = test_imgs.astype(float) / 255
```

```python
# now lets build a simple artificial neural network to learn from this data before the CNN
ann = Sequential()
ann.add(Flatten(input_shape=(32, 32, 3)))
ann.add(Dense(1000, activation='relu'))
ann.add(Dense(500, activation='relu'))
ann.add(Dense(10, activation='sigmoid'))

ann.compile(
    optimizer='SGD', 
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
ann.fit(X_train, train_labels, epochs=5)
```
